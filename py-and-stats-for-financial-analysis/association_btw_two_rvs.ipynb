### Association between 2 Random Variables

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import the housing information for analysis

housing = pd.read_csv('data/housing.csv', index_col=0)
housing.head()
```

```python
# Use covariance to calculate the association

housing.cov()
```

```python
# Use correlation to calculate the association is more appropriate in this case
housing.corr()
```

```python
# scatter matrix plot
from pandas.plotting import scatter_matrix
sm = scatter_matrix(housing, figsize=(15,15))
```

### Observe the association between LSTAT & MEDV

```python
# This time we take a closer look at MEDV vs LSTAT. What is the association between MEDV & LSTAT you observed?
housing.plot(kind='scatter', x='LSTAT', y='MEDV', figsize=(10,10))
```
